```python
import pandas as pd
import numpy as np

df = pd.read_csv('marketing_customer_analysis.csv')

df_num = df.select_dtypes(include=np.number)

#display(df_num)

# X-y split

y = df_num['Total Claim Amount']
X = df_num.drop(['Total Claim Amount'], axis=1)

X.head()
```

```python
# Normalized numericak

from sklearn.preprocessing import StandardScaler

Standardtransformer = StandardScaler()

Standardtransformer.fit(X) 

x_standardized = Standardtransformer.transform(X) 

x_standardized = pd.DataFrame(x_standardized,columns=X.columns)

round(x_standardized.head(),2)
```
